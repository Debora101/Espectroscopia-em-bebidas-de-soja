# PCA
O PCA é um metodo de redução de dimensionalidade quando temos muitas variáveis em nosso dataset.


Primeiramente vamos normalizar os dados antes de aplicar o PCA. Isso é importante porque o PCA é afetado pelas escalas das variáveis. Normalizamos os dados para que tenham média 0 e desvio padrão 1.

```python
from sklearn.preprocessing import StandardScaler

# Separando as variáveis independentes e a variável dependente
X = df.drop('GlicídiosTo', axis=1)
y = df['GlicídiosTo']

# Normalizando as variáveis independentes
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Os dados foram centrados na média e normalizados. Agora, vamos realizar a análise de componentes principais (PCA). O objetivo é encontrar a quantidade de variação que cada componente principal explica e para encontrar o número ideal de componentes principais. Para isso, inicialmente vamos calcular o PCA sem especificar o número de componentes para que possamos observar a variação explicada por cada componente. Depois disso, podemos decidir sobre o número ideal de componentes principais.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Criando o objeto PCA
pca = PCA()

# Ajustando e transformando os dados normalizados
X_pca = pca.fit_transform(X_scaled)

# Calculando a variância explicada cumulativa
explained_variance_ratio_cumsum = np.cumsum(pca.explained_variance_ratio_)

# Plotando a variância explicada por cada componente principal
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio_cumsum) + 1), explained_variance_ratio_cumsum, marker='o', linestyle='--')
plt.title('Variância Explicada Cumulativa dos Componentes Principais')
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Explicada Cumulativa')

plt.show()
```

O gráfico acima mostra a variância explicada cumulativa em função do número de componentes principais. A linha pontilhada representa a proporção acumulada da variância total que é explicada pelos componentes principais.

Podemos ver que a variância explicada aumenta acentuadamente para os primeiros componentes e depois se estabiliza, o que sugere que podemos capturar a maioria das informações nos dados com um número relativamente pequeno de componentes.

Vamos encontrar o número de componentes que explica pelo menos 95% da variância.

```python
# Encontrando o número de componentes que explica pelo menos 95% da variância
n_components_95 = np.where(explained_variance_ratio_cumsum >= 0.95)[0][0] + 1  # Adicionando 1 porque a indexação começa em 0

n_components_95
```

Agora, vamos realizar o PCA novamente, mas desta vez especificando o número de componentes como 8. Em seguida, vamos plotar os dois primeiros componentes principais. Estes são os dois componentes que explicam a maior parte da variância nos dados. O gráfico nos dará uma visão geral de como nossas amostras estão distribuídas no espaço definido pelos dois primeiros componentes principais.



```python
# Criando o objeto PCA com 8 componentes
pca = PCA(n_components=n_components_95)

# Ajustando e transformando os dados normalizados
X_pca = pca.fit_transform(X_scaled)

# Plotando os dois primeiros componentes principais
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('Distribuição das Amostras nos Dois Primeiros Componentes Principais')
plt.xlabel('Primeiro Componente Principal')
plt.ylabel('Segundo Componente Principal')

plt.show()
```

O gráfico acima representa a distribuição das amostras no espaço definido pelos dois primeiros componentes principais. Cada ponto no gráfico corresponde a uma amostra do conjunto de dados.

Os dois primeiros componentes principais são as direções no espaço de características que maximizam a variância nos dados. Assim, eles são as direções ao longo das quais os dados são mais espalhados, e portanto, são as direções que capturam a maior parte da informação nos dados.

Com base neste gráfico, podemos ver que as amostras estão bastante espalhadas no espaço definido pelos dois primeiros componentes principais. Isso sugere que esses dois componentes são eficazes para capturar a variabilidade nos dados.

Em conclusão, ao reduzir a dimensionalidade do nosso conjunto de dados para 8 componentes principais, conseguimos reter 95% da variância (informação) originalmente presente nos dados. Isso pode facilitar a análise subsequente, tornando os cálculos mais eficientes e os resultados mais fáceis de interpretar, ao mesmo tempo em que mantém a maior parte da informação útil contida nos dados

```python
# Plot the explained variance ratio for each component
plt.figure(figsize=(8, 5))
plt.bar(range(1, 9), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.xlabel('Componente Principal')
plt.ylabel('Variância Explicada (%)')
plt.title('Variância Explicada por Cada Componente Principal')
# Add the explained variance ratio on top of each bar
for i, v in enumerate(pca.explained_variance_ratio_):
    plt.text(i+1, v + 0.01, f"{v*100:.2f}%", ha='center')

plt.show()
```

O gráfico acima mostra a proporção da variância explicada por cada um dos 8 componentes principais. O valor acima de cada barra representa a porcentagem da variância total dos dados que é explicada por esse componente.

# Suavização Savitzky-Golay
À esquerda, temos os espectros originais, e à direita, os espectros após a aplicação do filtro Savitzky-Golay.

Esse código aplica o filtro Savitzky-Golay aos dados originais X, usando uma janela de 13 pontos e ajustando um polinômio de primeira ordem aos dados dentro dessa janela. O deriv=1 indica que estamos computando a primeira derivada dos dados, o que pode ajudar a realçar as características mais importantes do espectro.



```python
import numpy as np
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt

# Plotting the original and smoothed data side by side with the smoothing process included
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# Extract response variable
y = df['GlicídiosTo'].values

# Extract absorbance signals
X = df.values[:, 1:]

# Normalizing the independent variables
wl = np.linspace(1899, 900, num=X.shape[1], endpoint=True)

# Plot original data
axes[0].plot(wl, X.T)
axes[0].set_xlabel("Comprimento de onda (cm-1)")
axes[0].set_ylabel("Absorbance")
axes[0].set_xlim(reversed(axes[0].get_xlim()))  # Reverse the x-axis
axes[0].set_title("Original Spectra")

# Applying the Savitzky-Golay filter
X2 = savgol_filter(X, 13, polyorder=1, deriv=1)

# Plot smoothed data
axes[1].plot(wl, X2.T)
axes[1].set_xlabel("Comprimento de onda (cm-1)")
axes[1].set_ylabel("Absorbance")
axes[1].set_xlim(reversed(axes[1].get_xlim()))  # Reverse the x-axis
axes[1].set_title("Smoothed Spectra")

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

# Initialize PCA model
pca = PCA()

# Fit PCA on smoothed data
pca.fit(X2)

# Get the explained variance ratios
explained_variance = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance)


```

```python
# Plotting the first two principal components
plt.figure(figsize=(5,3))
plt.scatter(X_pca_95[:, 0], X_pca_95[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Smoothed Data')
plt.grid(True)
plt.show()
```

```python
# Plotting the explained variance for each principal component in percentages
plt.figure(figsize=(8, 5))
bars = plt.bar(range(1, len(pca_95.explained_variance_ratio_)+1), pca_95.explained_variance_ratio_ * 100, alpha=0.5, align='center')
plt.xlabel('Componente Principal')
plt.ylabel('Variância Explicada (%)')
plt.title('Variância Explicada por Cada Componente Principal')

# Adding the individual explained variances on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom')

plt.show()
```

# PCR
O PCR (Principal Component Regression) foi aplicado aos dados. Esse método combina o PCA para redução de dimensionalidade com a regressão linear para a predição. Nesse caso, usamos os 6 primeiros componentes principais para a predição (PCA com dados suavizados).

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=42)

# Create a PCA object
pca = PCA(n_components=6) # As we found that 6 components explain 95% variance

# Create a linear regression object
lm = LinearRegression()

# Create a pipeline of PCA and linear regression
model = make_pipeline(pca, lm)

# Train the model
model.fit(X_train, y_train)

# Predict the test set results
y_pred = model.predict(X_test)

# Calculate the mean squared error of the predictions
mse = mean_squared_error(y_test, y_pred)

mse
from sklearn.metrics import r2_score

# Calculate the R^2 (coefficient of determination) of the predictions
r2 = r2_score(y_test, y_pred)
print (r2,mse)
```

A métrica usada para avaliar o modelo foi o erro quadrático médio (MSE), que é uma medida comum para avaliar a precisão de modelos de regressão. O MSE calculado foi de aproximadamente 0.085.

O coeficiente de determinação R^2 para as previsões no conjunto de testes é aproximadamente 0.953.

O valor de R^2  varia entre 0 e 1, sendo que um valor de 1 indica que o modelo é capaz de explicar toda a variabilidade dos dados de resposta em torno de sua média. Portanto, um R^2 de 0.953 significa que o modelo é capaz de explicar aproximadamente 95.3% da variância na variável resposta "GlicídiosTo".

# PLS
O PLS é especialmente útil quando temos um grande número de preditores que são correlacionados entre si (multicolinearidade), e quando esses preditores são mais numerosos do que as observações. A PLS combina características da análise de componentes principais (PCA) e da regressão linear para criar um modelo preditivo.

Uma parte muito importante é separar nosso dataset em dois subconjuntos, um para calibrar o modelo e outro para validá-lo.

Para isso, vamos usar a função train_test_split passando como argumento as nossas variáveis x e y. Essa função divide randomicamente o conjunto de dados utilizado e retorna 4 vetores: os subconjuntos de variáveis independentes para treino e teste e de variáveis dependentes para treino e teste. Essa divisão é feita com base no argumento test_size=0.33 que foi passado. Isso quer dizer que **⅓ do conjunto de dados vai ser utilizado para testar**(27 amostras) o modelo e **⅔ para treiná-lo**(53 amostras). Além disso, também incluímos um random_state=123. Isso serve basicamente para que possamos ter o mesmo subconjunto de dados nas próximas vezes que utilizarmos o algoritmo. Isso é importante para a reprodutibilidade dos resultados.

Depois de dividir o nosso conjunto de dados, construímos o  modelo PLSRegression.Para treinar o modelo, vamos usar o método .fit(), passando como argumento o nosso subconjunto de treino.

Depois de treinarmos o modelo, podemos testar sua eficiência para novos dados. Porém, uma estratégia muito boa para se seguir antes de tentar validar o modelo com novos dados, é tentar mensurar a sua eficiência utilizando os mesmos dados que usamos para calibrá-lo. Isso é chamado de validação cruzada (cross-validation). Se o resultado for ruim nessa etapa, dificilmente será melhor quando os espectros de teste forem usados. A biblioteca scikit learn já possui um módulo para isso, o cross_val_predict. Para usá-lo precisamos informar o modelo (PLSRegression), e os espectros que usamos para treinar (x_train e y_train). Além disso, também é necessário o argumento cv. Esse argumento diz respeito a estratégia de divisão do conjunto de amostras. Ao passarmos o valor de 10, isso quer dizer que nosso conjunto de dados será dividido em 10 grupos de N amostras. Dessas N amostras 1 será utilizada para validar o modelo e N -1 amostras serão usadas para treiná-lo. Este processo será repetido N vezes.

```python
from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.33, random_state=123)

# Fit the PLS model on the training data with 6 components
pls = PLSRegression(n_components=6)
pls.fit(X_train, y_train)

# Perform cross-validation on the training data
y_cv = cross_val_predict(pls, X_train, y_train, cv=10)

# Calculate scores for the cross-validation
mse_cv = mean_squared_error(y_train, y_cv)
r2_cv = r2_score(y_train, y_cv)

# Predict the response variable for the test data
y_pred = pls.predict(X_test)

# Calculate scores for the test data
mse_test = mean_squared_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)

mse_cv, r2_cv, mse_test, r2_test

```

 Erro Quadrático Médio (MSE): 0.0583

2. Para o conjunto de teste:

 Erro Quadrático Médio (MSE): 0.0281

 Coeficiente de Determinação (R²): 0.9688

# RLM
A regressão linear múltipla é uma técnica estatística que usa várias variáveis explicativas para prever o resultado de uma variável resposta. O objetivo da regressão linear múltipla é modelar a relação linear entre as variáveis explicativas (independentes) e a variável resposta (dependente).

A regressão linear múltipla é representada pela equação:

image.png

onde:

°  y é a variável de resposta

°  X1,X2,.... Xn são as variáveis explicativas

°  b0 é o intercepto do modelo

°  b1,b2,...bn são os coeficientes de regressão, que representam o efeito que cada variável explicativa tem na variável de resposta.

°  *e* é o erro, que é a diferença entre o valor previsto e o valor real da variável de resposta.

O objetivo da regressão linear múltipla é encontrar os coeficientes de regressão que minimizam a soma dos quadrados dos resíduos (ou seja, a diferença entre os valores reais e previstos da variável de resposta).


```python
# Usando dados suavizados.
#Split the data into training and test sets with 2/3 of data for training and 1/3 for testing
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.33, random_state=123)

# Create a Linear Regression object
lm = LinearRegression()

# Fit the model to the training data
lm.fit(X_train, y_train)

# Perform cross-validation on the training data
y_cv = cross_val_predict(lm, X_train, y_train, cv=10)

# Predict the response variable for the test data
y_test_pred = lm.predict(X_test)

# Calculate the performance metrics for the training data (cross-validation)
mse_train = mean_squared_error(y_train, y_cv)
r2_train = r2_score(y_train, y_cv)

# Calculate the performance metrics for the test data
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

mse_train, r2_train, mse_test, r2_test
```

# REGRESSÃO LASSO com LASSOCV
A Regressão Lasso (Least Absolute Shrinkage and Selection Operator) é um tipo de análise de regressão que realiza tanto a seleção de variáveis quanto a regularização, a fim de melhorar a precisão e a interpretabilidade do modelo estatístico produzido.

A principal característica da regressão Lasso é que ela pode reduzir alguns coeficientes de regressão para exatamente zero. Isso significa que a Lasso não apenas ajuda a evitar o overfitting, mas também realiza a seleção de variáveis.

Em um modelo de regressão, cada variável de entrada (ou característica) é associada a um coeficiente que quantifica sua relação com a variável de saída.

No entanto, na regressão Lasso, à medida que o valor de alpha aumenta, mais e mais coeficientes de características se tornam exatamente zero. Isso significa que essas características são efetivamente excluídas do modelo. Isso pode ser extremamente útil quando você tem um grande número de características e acredita que apenas um subconjunto delas realmente importa.

**Aqui está um resumo do que foi feito no código com o Lasso**:

**1° Estimativa do melhor valor de alpha com LassoCV:** Foi feito ums validação cruzada cv=10. O objetivo disso é encontrar o valor de alpha que minimiza o erro médio em todas as iterações da validação cruzada. Ao especificar alphas=None, a função LassoCV irá testar uma gama de valores de alpha e escolher o mais adequado.

**2° Aplicação da Regressão Lasso com o alpha ótimo:** Após a obtenção do melhor valor de alpha, a função LassoCV foi utilizada novamente para ajustar um modelo de Regressão Lasso com este valor específico de alpha. Neste caso, foi especificado alphas=[best_alpha].

**3° Previsão e cálculo de métricas de desempenho:** Em seguida, o modelo ajustado foi utilizado para prever os valores da variável dependente (y) tanto para os dados de treinamento quanto para os dados de teste

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.33, random_state=42)
# Perform Lasso Regression with cross-validation to find the optimal alpha
lasso_cv = LassoCV(alphas=None, cv=10, max_iter=10000)

# Fit the model to the training data
lasso_cv.fit(X_train, y_train)

# Get the best alpha value
best_alpha = lasso_cv.alpha_

# Now fit a lasso model with the optimal alpha
lasso_optimal = LassoCV(alphas=[best_alpha], cv=10, max_iter=10000)
lasso_optimal.fit(X_train, y_train)

# Predict the values
y_train_pred_opt = lasso_optimal.predict(X_train)
y_test_pred_opt = lasso_optimal.predict(X_test)

# Calculate MSE and R^2 for training data
mse_train_opt = mean_squared_error(y_train, y_train_pred_opt)
r2_train_opt = r2_score(y_train, y_train_pred_opt)

# Calculate MSE and R^2 for testing data
mse_test_opt = mean_squared_error(y_test, y_test_pred_opt)
r2_test_opt = r2_score(y_test, y_test_pred_opt)

best_alpha, mse_train_opt, r2_train_opt, mse_test_opt, r2_test_opt
```

O melhor valor de alpha encontrado foi
7.47X10^-7 ( 0.0000007472).

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define a set of possible alpha values
alphas = np.logspace(-4, -0.5, 30)

# Perform cross-validation to choose the best alpha
lasso_cv = LassoCV(alphas=alphas, cv=5, random_state=0, max_iter=10000)
lasso_cv.fit(X_train, y_train)

# Get the best alpha
best_alpha = lasso_cv.alpha_

best_alpha
```

# **Curva de aprendizado das Regressões Lasso**

Uma abordagem comum é usar a curva de aprendizado, que mostra o desempenho do modelo nos conjuntos de treinamento e validação à medida que o tamanho do conjunto de treinamento aumenta. Se o modelo está overfitting, podemos esperar ver uma grande lacuna entre as curvas de treinamento e validação, porque o modelo está se ajustando muito bem aos dados de treinamento, mas não está generalizando bem para os dados de validação.

Vamos gerar as curvas de aprendizado para ambos os modelos Lasso que treinamos. O segundo com alpha α≈0.00646 e o primeiro com α ≈ 0.0000007472.

Para isso, vamos usar a função learning_curve do scikit-learn, que gera as curvas de aprendizado automaticamente.


```python
from sklearn.model_selection import learning_curve
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

def plot_learning_curve(model, X, y, ax=None):
    """Plot the learning curve of a model"""
    train_sizes, train_scores, valid_scores = learning_curve(model, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 5))
    train_scores_mean = np.mean(train_scores, axis=1)
    valid_scores_mean = np.mean(valid_scores, axis=1)

    if ax is None:
        fig, ax = plt.subplots()

    ax.plot(train_sizes, train_scores_mean, label='Training score')
    ax.plot(train_sizes, valid_scores_mean, label='Cross-validation score')
    ax.legend()
    ax.set_xlabel('Training set size')
    ax.set_ylabel('Score')
    ax.set_title('Learning curve')

# Define the models
lasso1 = make_pipeline(StandardScaler(), Lasso(alpha=0.00646, max_iter=10000))
lasso2 = make_pipeline(StandardScaler(), Lasso(alpha=7.47e-7, max_iter=10000))

# Plot the learning curves
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

plot_learning_curve(lasso1, X2, y, ax=axes[0])
axes[0].set_title('Learning curve (Lasso, alpha=0.00646)')

plot_learning_curve(lasso2, X2, y, ax=axes[1])
axes[1].set_title('Learning curve (Lasso, alpha=7.47e-7)')

plt.tight_layout()
plt.show()
```

Ambos os modelos parecem ter bom desempenho e não apresentam sinais significativos de overfitting, com base nas curvas de aprendizado e nas métricas que calculamos.

O modelo com α≈0.00646 tem um valor de
α maior o que significa que ele usa uma forma mais forte de regularização e, portamente, tem uma estrutura mais simples. Isso pode torná-lo mais robusto a pequenas variações nos dados e melhorar a sua capacidade de generalização para novos dados. Além disso, este modelo teve um desempenho ligeiramente melhor no conjunto de teste em termos de MSE e R^2.


Por outro lado, o modelo α≈ 7.47x10^-7 tem um valor de α  menor, o que significa que ele usa uma forma mais fraca de regularização e pode capturar mais detalhes nos dados. Isso pode torná-lo mais adequado se queremos capturar as nuances mais finas dos  dados.
